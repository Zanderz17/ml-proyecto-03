# Librerias

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

# PCA & r21d

```python
csv_path = './procesed-data/full/r21d.csv'
data_r21d = pd.read_csv(csv_path)

csv_path = './procesed-data-validation/full/r21d.csv'
data_r21d_validation = pd.read_csv(csv_path, header=0)

csv_path = './procesed-data-testing/full/r21d.csv'
data_r21d_testing = pd.read_csv(csv_path, header=0)

data_r21d_validation = data_r21d_validation.iloc[:, :-1]
data_r21d_testing = data_r21d_testing.iloc[:, :-1]
```

```python
data_r21d.shape
```

```python
data_r21d_testing.columns = data_r21d.columns
data_r21d_validation.columns = data_r21d.columns
```

## Eliminando NaN

```python
# Eliminar filas que contienen cualquier NaN
data_cleaned_r21d = data_r21d.dropna()

print(f'Número original de filas: {len(data_r21d)}')
print(f'Número de filas después de eliminar NaN: {len(data_cleaned_r21d)}')
```

```python
# Eliminar filas que contienen cualquier NaN
data_cleaned_r21d_validation = data_r21d_validation.dropna()

print(f'Número original de filas: {len(data_r21d_validation)}')
print(f'Número de filas después de eliminar NaN: {len(data_cleaned_r21d_validation)}')
```

```python
# Eliminar filas que contienen cualquier NaN
data_cleaned_r21d_testing = data_r21d_testing.dropna()

print(f'Número original de filas: {len(data_r21d_testing)}')
print(f'Número de filas después de eliminar NaN: {len(data_cleaned_r21d_testing)}')
```

```python
# PCA
pca = PCA(n_components=1000)  
data_r21d_pca = pca.fit_transform(data_cleaned_r21d)

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print(cumulative_explained_variance[-1])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance * 100, 'b', label='Cumulative explained variance')
plt.axhline(y=90, color='r', linestyle='--', label='90% Explained Variance')
plt.ylabel('Explained variance ratio (%)')
plt.xlabel('Principal component')
plt.legend(loc='best')
plt.title('Explained Variance of PCA Components')
plt.grid(True)
plt.show()
```

```python
data_r21d_validation_pca = pca.transform(data_cleaned_r21d_validation)
```

```python
data_r21d_testing_pca = pca.transform(data_cleaned_r21d_testing)
```

## Guardar PCA-r21d en CSV

```python
output_csv_path = './procesed-data/reduced/pca/r21d-cleaned.csv'
np.savetxt(output_csv_path, data_r21d_pca, delimiter=',', header='', comments='')
```

```python
output_csv_path = './procesed-data-validation/reduced/pca/r21d-cleaned.csv'
np.savetxt(output_csv_path, data_r21d_validation_pca, delimiter=',', header='', comments='')
```

```python
output_csv_path = './procesed-data-testing/reduced/pca/r21d-cleaned.csv'
np.savetxt(output_csv_path, data_r21d_testing_pca, delimiter=',', header='', comments='')
```
